# Get some data and plot it!

Load excel data and plot average temperature by month in one country.

Uses [pandas](https://pandas.pydata.org/), an open source library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming language.

## Download dataset from the Climate Change Knowledge portal
1.	Visit http://sdwebx.worldbank.org/climateportal/index.cfm
2.	Click on the area, then country of your interest on the interactive map
3.	Click the ‘Click to download historical data’ link
4.	Select ‘Temperature’, the country and the time period you are interested in
5.	Click ‘Download Data To Excel’, and save the file on your computer as temperatures.xls
6.	Select ‘Rainfall’, the country and the time period you are interested in
7.	Click ‘Download Data To Excel’, and save the file on your computer as rainfall.xls

````python
import pandas as pd

# This could be an URL instead (but the climate portal does not let you select a country)
raw_temp = pd.read_excel('temperatures.xls',
                         # This is to avoid strange characters in the file such as "\t" in Year
                         names=['tas', 'year', 'month', 'country', 'ISO3', 'ISO2'])
        
raw_temp.keys()
````

## Take a look at the data

````python
raw_temp
````

## Get some description

````python
raw_temp.describe()
````

## We are interested in the temperature average per month

````python
monthly_temp = raw_temp.groupby('month')['tas']

monthly_temp.describe()
````

````python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

monthly_temp.mean().plot()
````

## Average year temperature

````python
yearly_temp = raw_temp.groupby('year')['tas']

yearly_temp.mean().plot()
````
